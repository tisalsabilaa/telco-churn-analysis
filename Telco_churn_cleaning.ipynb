```python
import pandas as pd
```

```python
df = pd.read_csv('Telco_Customer_Data.csv')

print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
print(df['TotalCharges'].unique()[:20])
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
print(df['TotalCharges'].isnull().sum())
print(df[df['TotalCharges'].isnull()])
```

```python
df = df.dropna(subset=['TotalCharges'])
print(df['TotalCharges'].dtype)

df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
df['Churn_Binary'] = df['Churn'].map({'No':0, 'Yes':1})

print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
print(df[['tenure', 'MonthlyCharges', 'TotalCharges']].describe())
```

```python
df.to_csv('telco_cleaned.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
churn_rate = df['Churn_Binary'].mean() * 100
print(f"Overall Churn Rate: {churn_rate:.2f}%")

print(df.groupby('Contract')['Churn_Binary'].mean().sort_values(ascending=False))

print(df.groupby('InternetService')['Churn_Binary'].mean().sort_values(ascending=False))

df['Tenure_Group'] = pd.cut(df['tenure'], bins=[0,12,24,48,72], 
                            labels=['0-12mo', '13-24mo', '25-48mo', '49-72mo'])
print(df.groupby('Tenure_Group', observed=True)['Churn_Binary'].mean().sort_values(ascending=False))

print(df.groupby('SeniorCitizen')['Churn_Binary'].mean())
```

```python
remaining = ['PaymentMethod', 'PaperlessBilling', 'TechSupport', 
             'OnlineSecurity', 'StreamingTV', 'StreamingMovies']

for col in remaining:
    print(f"\nChurn by {col}:")
    print(df.groupby(col)['Churn_Binary'].mean().sort_values(ascending=False))
```
